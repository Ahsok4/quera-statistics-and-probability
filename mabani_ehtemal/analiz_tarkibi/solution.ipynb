اصول شمارش
        </font>
        <br>
        فرض کنید بتوان از شهر A از طریق n جاده به شهر B رفت و از شهر B از طریق m جاده به شهر C رفت. اگر قرار باشد با عبور از B؛ از A به C سفر کنیم؛ آنگاه به ازای هر راهی که برای سفر از A بهB انتخاب می‌کنیم برای سفر از B به C ؛m انتخاب داریم. در نتیجه در کل برای مسافرت از A به C از طریق B؛ mn انتخاب وجود دارد. این مثال ساده ای از اصل ضرب است که تعریف آن در ادامه آمده.
		<p></p>
		<hr>
         اصل ضرب: فرض کنید نحوه انجام کاری را بتوان به k مرحله تجزیه کرد. مرحله اول را به $n_1$ طریق بتوان انجام داد، مرحله دوم را به $n_2$ طریق بتوان انجام داد،... و مرحله k ام به $n_k$ طریق قابل انجام باشد. در این صورت کل کار را به $n_1\times n_2 \times ... \times n_k$ طریق می توان انجام داد.
         برای مثال اگر بخواهیم تعداد اعداد دوررقمی فرد را پیدا کنیم می توان گفت برای نوشتن یک عدد دورقمی ابتدا رقم دهگان و سپس رقم یکان آن را بنویسیم. ۹ انتخاب برای رقم دهگان و به ازای هر انتخاب رقم دهگان ۵ انتخاب(۱و۳و۵و۷و۹) برای رقم یکان داریم پس تعداد اعداد دورقمی فرد می شود: 
        $9\times 5 = 45$
        <br/>
        <br/>
        اصل جمع: اگر برای انجام کارA لازم باشد کار های $A_1 , A_2 , ... , A_n$ را انجام داد و هیچ دو تایی از آن ها اشتراک نداشته باشند و کار $A_1$ را به $n_1$ طریق، کار $A_2$ را به $n_2$ راه مختلف، ... و کار $A_n$ را به $n_n$ طریق مختلف بتوان انجام داد آنگاه کار A را به $n_1 + n_2 + ... + n_n$ طریق مختلف می توان انجام داد.
        برای مثال برای به دست آوردن تعداد اعداد دورقمی زوج با ارقام متمایز می‌توان گفت برای رقم دهگان ۹ انتخاب داریم. تعداد انتخاب های رقم یکان بستگی به این دارد که رقم دهگان انتخاب شده زوج باشد یا فرد. پس مساله را به دوحالت بدون اشتراک تقسیم کرد: 
        ۱- رقم دهگان زوج باشد. در این حالت ۴ انتخاب برای رقم‌های دهگان و ۴ انتخاب برای یکان و طبق اصل ضرب ۱۶ عدد داریم. 
        ۲- رقم دهگان فرد باشد که در این حالت تعداد انتخاب های دهگان و یکان به ترتیب ۵ و۵ است که باز طبق اصل ضرب می‌شود ۲۵ 
          و در مجموع می‌شود: 
$16+25 = 41$
        <br/>
        <br/>
        اصل متمم: فرض کنید B زیر مجموعه‌ای از مجموعه A باشد، در این صورت: $|A - B| = |A| - |B|$
         مثلا اگر بخواهیم همان مثال تعداد اعداد دورقمی زوج با ارقام متمایز که با اصل جمع حل کردیم را با کمک اصل متمم بدست آوریم می‌توان گفت اکر تعداد اعداد فرد دورقمی با رقم متمایز را داشته باشیم(۴۰ تا) تعداد اعداد زوج دورقمی با رقم متمایز می‌شود: تعداد همه اعداد دورقمی با اراقام متمایز منهای تعداد اعداد فرد دورقمی با ارقام متمایز یعنی: 
         $81-40 = 41$

        <br/>
        جایگشت r شی از n شی: اگر مجموعه‌ای از n شی در اختیار داشته باشیم، هر آرایش خطی متشکل از r تا از این اشیا را یک جایگشت r شی از این n شی می‌نامیم. 
        <br/>
        <br/>
        به جایگشت r از n تبدیل r از n نیز می‌گویند و آن را با نماد $P(n , r)$ یا $P_r^n$ می‌دهند.
        <br/>
        مثال: با حروف $A, B, C, D$ چند کلمه ۴ حرفی می‌توان ساخت؟

بیایید در ابتدا فرض کنیم این اشیا همه متمایز اند. پاسخ می‌شود $۸!$. حال هر کدام از عروسک های مشابه اگر متمایز بودند  چند جایگشت داشتند؟
 سگ $۳!$ و گربه $۲!$
این حالات در واقع نباید در محاسبه اولیه ما می‌بوده اند چون تکراری هستند و حالت جدیدی تولید نمی‌کنند(به عبارت دیگر نتیجه جابه جا کردن ترتیب قرارگیری اشیا مشابه را نباید جایگشت جدید به حساب آوریم). بنابر اصل ضرب این حالات قبلا در حاصل کل ضرب شده اند، پس کافی است ما 
حاصل قبلی را تقسیم بر تعداد حالات مربوط به موارد تکراری کنیم:
        $\frac{۸!}{۲! ۳!}$
<br>
        <br/>
        بصورت کلی
        n شی در نظر بگیرید طوری که $k_1$ شی آن با یک دیگر، $k_2$ شی دیگر با هم، ... و $k_t$ شی دیگر با هم مشابه باشند طوری که $k_1 + k_2 + ... + k_t = n$ شود. در آن صورت تعداد جایگشت‌های متمایز n شی برابر است با:
        $$p(n;k_1,k_2,.....,k_t)=\frac{n!}{k_1! k_2! ... k_t!}$$
        <br>
        معمولا در کتاب ها ترتیب(جایشگت) را با علامت p نمایش می‌دهند. این p را با  P به معنی احتمال که به زودی معرفی خواهیم کرد اشتباه نکنید!

       <br>
       تعریف ترکیب: اگر A مجموعه باشد، یک ترکیب A یعنی زیر مجموعه‌ای از A. به بیان دیگر یک آرایش نامرتب r شی از مجموعه A که شامل n شی است یک ترکیب r از n می‌گوییم.
        <br/>
        تعداد زیر مجموعه‌های r عضوی از مجموعه‌ای n عضوی(ترکیب r از n) را با نماد $n \choose r$ نمایش می‌دهیم.
        <br/>
        اگر $ 0 \leq r \leq n$ باشد، آنگاه:
        ${n \choose r} = \frac{n!}{r!(n - r)!}$
        <br/>
        مثال: n نقطه روی محیط دایره انتخاب می‌کنیم. همه وتر‌های بین هر جفت از این نقاط را رسم می‌کنیم. با فرض این که هیچ 3 وتری در یک نقطه همرس نیستند تعداد نقاط تقاطع وتر‌ها درون دایره را بیابید.
        <br/>
        هر 4 نقطه ار بین n نقطه یک نقطه تقاطع متمایز تولید می‌کنند و هر نقطه تقاطع از تقاطع دو وتر تشکیل می‌شود که انتهای دو وتر 4 نقطه متمایز از n نقطه است. پس تعداد نقاط تقاطع برابر است با: $n \choose 4$
